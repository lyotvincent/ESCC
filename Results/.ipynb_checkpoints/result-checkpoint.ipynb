```python
for elem1 in ['SC3','SAFE','ESCC','ESCC+']:
    for elem2 in ['Biase','Goolam','Deng','Deng(another)','Kolo','Pollen','Yan']:
        try:
            f=open(elem1+'_'+elem2+'.txt')
        except:
            continue
        text=f.read()
        if elem1=='SC3' or elem1=='SAFE':
            cluster=np.array([elem.split(' ') for elem in text.split('\n')[:-1]],dtype=np.int)
        else:
            cluster=np.array([elem.split(',')[:-1] for elem in text.split('\n')[:-1]],dtype=np.int)
        from sklearn import metrics
        for elem in cluster:
            print(elem1,elem2,metrics.adjusted_rand_score(color[elem2],elem))
        #print(elem1,elem2,np.mean([metrics.adjusted_rand_score(color[elem2 if elem2!='Deng(another)' else 'Deng'],elem) for elem in cluster]))
        print('end')
```

```python
import scipy
from sklearn.preprocessing import normalize
def Lap(K):
    tp=np.sum(K,1)
    D=np.mat(np.diag(tp))
    d=np.mat(np.diag(1/np.sqrt(tp)))
    L=d*(D-K)*d
    u,v=scipy.linalg.eigh(L)
    rank=np.argsort(u)
    return np.array(normalize(v.T[rank[1:]].real, norm='l2')).transpose()
```

```python
res=list()
for elem in ['Biase','Goolam','Deng','Kolo','Pollen','Yan']:
    f1,f2=open(elem+'_clusters1.txt'),open(elem+'_clusters2.txt')
    text1,text2=f1.read(),f2.read()
    cluster1=np.array([elem.split(',')[:-1] for elem in text1.split('\n')[:-1]],dtype=np.int)
    cluster2=np.array([elem.split(',')[:-1] for elem in text2.split('\n')[:-1]],dtype=np.int)
    tp=list()
    for i in range(100):
        c1,c2=cluster1[i<<7:(i+1)<<7],cluster2[i<<6:(i+1)<<6]
        ARI1,ARI2=[metrics.adjusted_rand_score(color[elem],e) for e in c1],[metrics.adjusted_rand_score(color[elem],e) for e in c2]
        Lcoord1=Lap([[metrics.adjusted_rand_score(elem1,elem2) for elem2 in c1] for elem1 in c1])
        Lcoord2=Lap([[metrics.adjusted_rand_score(elem1,elem2) for elem2 in c2] for elem1 in c2])
        rela1=[scipy.stats.spearmanr(list(range(len(elem))),abs(elem))[0] for elem in Lcoord1]
        rela2=[scipy.stats.spearmanr(list(range(len(elem))),abs(elem))[0] for elem in Lcoord2]
        tp.append((np.mean(ARI2)-np.mean(ARI1),scipy.stats.spearmanr(rela1,ARI1),scipy.stats.spearmanr(rela2,ARI2)))
    res.append(tp)
```

```python
def rearrange(c):
    d=dict()
    re=list()
    for elem in c:
        if elem not in d.keys():
            d[elem]=len(d)
        re.append(d[elem])
    return re
```
